```python
page=requests.get('https://www.karar.com/yazarlar/zafer-acar') # Web scraping i engellemeyi kaldıran kodlar var
za_tum_soup=BeautifulSoup(page.content,'html.parser')
linkler=za_tum_soup.find_all('article',{'class':'item box-shadow'})
linkler
```

```python
#Tarihlerini al
for link in linkler:
    print(link.time.get('datetime'))
```

```python
for link in linkler:
    print('https://www.karar.com'+link.a.get('href'))
```

```python
basliklar=[]
yazilar=[]
for link in linkler:
    link1='https://www.karar.com'+link.a.get('href')
    #print(link.h3.text)
    baslik=link.h3.text
    page2=requests.get(link1)
    soup2=BeautifulSoup(page2.content,'html.parser')
    yazi=soup2.select('article.article-content')[0].text
    basliklar.append(baslik)
    yazilar.append(yazi)
```

```python
za_makaleler_df=pd.DataFrame({'Başlık':basliklar,'Yazı':yazilar})
```
